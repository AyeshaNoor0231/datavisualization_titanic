**filling the column of Age with Median Age of each class**

````python
df. head(20)
````

````python
print(df.isnull().sum())
````

````python
#Imputation function
def impute_age(cols):
  Age = cols[0]
  Pclass = cols[1]

  if pd.isnull(Age):

    if Pclass ==1:
      return 37
    elif Pclass == 2:
      return 29
    else:
      return 24

  else:
    return Age

#Calling the function

df['Age'] = df[['Age', 'Pclass']].apply(impute_age, axis = 1) #axis =1 column
````

````python
print(df.isnull().sum())
````

````python
df1 =df.drop(["Cabin", "Name",'Ticket', 'PassengerId'], axis =1, inplace= False) #inplace is modifing the original data, if false, will provided a new modified data, keeping the original one
df1
````

````python
df1
````

````python
print(df1.isnull().sum())
````

````python
df1.dropna(inplace=True)
````

````python
print(df1.isnull().sum())
````

````python
objcat = ['Sex', 'Embarked']
for col in objcat:
  df1[col] = df1[col].astype('category')
````

````python
#setting dummy values as o and 1 in place of category to make the entire data in numerical shape
sex = pd.get_dummies(df1['Sex'], drop_first=True)
Embarked = pd.get_dummies(df1['Embarked'],drop_first= True)

#Adding those values to the data set through concatenation
df1 = pd.concat([df1,sex, Embarked], axis = 1)

````

````python
df1.head(5)
````

````python
df1.drop(['Sex', 'Embarked'], axis = 1, inplace = True)
df1
````

Now our entire dataset is in numerical values and good to go to train a classification model through sklearn classification

````python
#we need to split input and output (x & y) from the dataset
x = df1.drop('Survived', axis = 1)
y = df1['Survived']

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.20, random_state = 101)
````

````python
#Training the model

#import model
from sklearn.linear_model import LogisticRegression

#Creating instance of model

lreg_m = LogisticRegression()

#Passing the data to the model to train
lreg_m.fit(x_train, y_train)
````

````python
#try the model by making predictions
y_pred_lreg_m = lreg_m.predict(x_test)
print(y_pred_lreg_m)
count_0 = 0
count_1 = 0

for i in y_pred_lreg_m:
    if i == 1:
        count_1 += 1
    elif i == 0:
        count_0 += 1

print("Count of Not-Survived:", count_0)
print("Count of Survived:", count_1)
print('total:',count_0 + count_1)
````

````python
#Check accuracy of the model trained
from sklearn.metrics import classification_report, accuracy_score
performance = accuracy_score(y_test, y_pred_lreg_m)
performance *100
````

# **Classification**

````python
pr =lreg_m.predict_proba(x_test)[:,1]
pr
````

````python
for i in range (len(y_test)):
  print(round(pr[i],3) , '--->' , y_pred_lreg_m[i])
````

# **Logistic Regression**

````python
# Score It
from sklearn.metrics import classification_report, accuracy_score

print('Classification Model')
# Accuracy
print('--'*30)
logreg_accuracy = round(accuracy_score(y_test, y_pred_lreg_m) * 100,2)
print('Accuracy', logreg_accuracy,'%')
````

# **Decision Tree**

````python
#Classification
from sklearn.tree import DecisionTreeClassifier

Dtree = DecisionTreeClassifier()
Dtree.fit(x_train, y_train)
````

````python
from sklearn.metrics import accuracy_score
y_pred_Dt = Dtree.predict(x_test)

Dt_acc = accuracy_score(y_test, y_pred_Dt) * 100
print(Dt_acc)
````

# **Random Forest**

````python
from sklearn.ensemble import RandomForestClassifier
Rf = RandomForestClassifier()
Rf.fit(x_train,y_train)
````

````python
y_Rf_pred = Rf.predict(x_test)

from sklearn.metrics import accuracy_score
Rf_acc = accuracy_score(y_test, y_Rf_pred)*100
Rf_acc
````

# **Gradien Boosting**

````python
from sklearn.ensemble import GradientBoostingClassifier

gb = GradientBoostingClassifier()
gb.fit(x_train, y_train)
````

````python
y_gb_pred = gb.predict(x_test)

from sklearn.metrics import accuracy_score

gb_acc = accuracy_score(y_test, y_gb_pred)*100
gb_acc
````

# **Ranking**

````python
# Sample model scores (replace these with your actual model scores)
model_scores = {
    "Logistic Regression": logreg_accuracy,
    "Decision Tree Classifier": Dt_acc,
    "Random Forest Classifier": Rf_acc,
    "Gradient Boosting Classifier": gb_acc
}

# Sort the model scores in descending order based on their values (higher values first)
sorted_scores = sorted(model_scores.items(), key=lambda x: x[1], reverse=True)

# Display the ranking of the models
print("Model Rankings (Greater Values are better):")
for rank, (model_name, score) in enumerate(sorted_scores, start=1):
    print(f"{rank}. {model_name}: {score}")
````
